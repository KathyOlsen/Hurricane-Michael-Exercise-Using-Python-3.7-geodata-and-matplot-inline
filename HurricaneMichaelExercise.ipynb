```python
import pandas as pd
import geopandas as gpd 
import matplotlib.pyplot as plt

from datetime import datetime  
from shapely.geometry import Point  
from geopandas.tools import geocode 
from geopandas.tools import reverse_geocode

%matplotlib inline
```

```python
Location = "datasets_hurricaneMichael.csv"
df = pd.read_csv(Location)
df.head()
```

```python
df.dtypes
```

```python
df.count()
```

```python
df['Longc'] = 0-df['Long']
df['Longc'].head(5)
```

```python
df['Coordinates'] = list(df[['Longc', 'Lat']].values)
df.head()
```

```python
df['Coordinates'] = df['Coordinates'].apply(Point)
df.head()
```

```python
geo_df = gpd.GeoDataFrame(df, geometry='Coordinates')
geo_df.head()
```

```python
type(geo_df['Coordinates'])
```

```python
MapLocation = "datasets_US_states(5m).json"
map_df = gpd.read_file(MapLocation)
map_df.head()
```

```python
map48 = map_df.loc[map_df['NAME'].isin(['Alaska','Hawaii','Puerto Rico'])==False]
map48.plot()
```

```python
def ms(x):
    
    if x <= 30: return 20
    if x > 30 and x <= 39: return 30
    if x > 39 and x <= 74: return 50
    if x > 74 and x <= 95: return 80
    if x > 95 and x <= 110: return 110
    if x > 110 and x <= 129: return 140
    if x > 129 and x <= 156: return 170
    if x > 156: return 210
    else: return 30
    

geo_df['ms']=geo_df['Wind'].apply(ms)

geo_df.head(45)
```

```python
fig, ax = plt.subplots(1, figsize=(15,10))
basemap = map48.plot(ax=ax)
geo_df.plot(ax=basemap, color='darkred', marker='o', markersize=geo_df['ms'])                                               
                                              
ax.axis('off')
ax.set_title("Hurricane Michael - October 6-12, 2018", fontsize=25, fontweight=3)
```
